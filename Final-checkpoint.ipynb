```python
## Converting the column matrix to row matrix of the mapping dataframe using MELT function

# store the value and id variables in two separate arrays

# store the value variables in one Series
value_vars = list(mapping.columns[1:])

# take the setdiff() to get the rest of the variables
id_vars = list(np.setdiff1d(mapping.columns, value_vars))

print(value_vars, "\n")
print(id_vars)
```

```python
# convert wide into long
mapping_long = pd.melt(mapping, 
        id_vars = id_vars, 
        value_vars = value_vars)
mapping_long
```

```python
# remove rows having value=0
mapping_long = mapping_long[mapping_long['value']==1]
mapping_long = mapping_long.drop('value', axis=1)
```
